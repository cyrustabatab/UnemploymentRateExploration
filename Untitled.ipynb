```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.offline as po
import json
%matplotlib inline
```

```python
import requests
```

```python
url = 'https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json'
```

```python
requests.get(url)
```

```python
county_geo_data =json.loads(requests.get(url).text)
```

```python
columns = ['state_fips_code','county_fips_code','county_name','year','labor_force','employed','unemployed','unemployment_rate']
```

```python
types = {'state_fips_code': 'str','county_fips_code': 'str'}
```

```python
countys = pd.read_excel("laucnty19.xlsx",usecols="B:E,G:J",skiprows=6,header=None,skipfooter=3,names=columns,dtype=types)
```

```python
countys
```

```python
countys['fip_code'] = countys['state_fips_code'] + countys['county_fips_code']
```

```python
countys
```

```python
fig = px.choropleth(countys, geojson=county_geo_data, locations='fip_code', color='unemployment_rate',
                           color_continuous_scale="Viridis",
                           range_color=(0,12),
                           scope="usa",
                            height=1000,
                            width=1000,
                           labels={'unemployment_rate':'unemployment_rate'}
                          )
po.plot(fig)
```
